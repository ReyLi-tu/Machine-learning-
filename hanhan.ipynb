```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
data = pd.read_csv("hw8.csv",header = None)
theta = data[1000]
data.drop(data.columns[len(data.columns)-1], axis=1, inplace=True)
data
```

```python
x1 = data.iloc[:, : 500].values
x2 = data.iloc[:,500 :].values
```

```python
def kac(theta):
    kac = (1/(2*np.pi))*(np.sin(theta)+(np.pi-theta)*np.cos(theta))
    return kac
```

```python
def approx(x1,x2,weight,n):
    sum=np.zeros(500)
    for index,val in enumerate(x1):
        for m in range (0,n):
            weight = s = np.random.normal(0, 1, 500)
            temp = np.multiply(np.maximum(0,np.dot(weight,x1[index,:])),(np.maximum(0,np.dot(weight,x2[index,:]))))
            sum[index] = sum[index]+temp
            
    return sum/n
```

```python
plt.plot(approx(x1,x2,weight,1000))
```
